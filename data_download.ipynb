```python
import os
from datasets import load_dataset
import json

# 保存先ディレクトリ
save_dir = "data/hf_datasets"
os.makedirs(save_dir, exist_ok=True)

# ダウンロード対象のリスト
dataset_ids = [
    "u-10bei/structured_data_with_cot_dataset_512_v2",
    "u-10bei/structured_data_with_cot_dataset_512_v4",
    "u-10bei/structured_data_with_cot_dataset_512_v5",
    "u-10bei/structured_data_with_cot_dataset_512",
    "u-10bei/structured_data_with_cot_dataset_v2",
    "u-10bei/structured_data_with_cot_dataset",
    "daichira/structured-3k-mix-sft",
    "daichira/structured-5k-mix-sft",
    "daichira/structured-hard-sft-4k",
    "u-10bei/dpo-dataset-qwen-cot"
]

def save_dataset_locally(dataset_id):
    name = dataset_id.split("/")[-1]
    save_path = os.path.join(save_dir, f"{name}.jsonl")
    
    print(f"📥 Downloading {dataset_id}...")
    try:
        # trainスプリットを読み込み
        ds = load_dataset(dataset_id, split="train")
        
        # JSONL形式で保存
        with open(save_path, "w", encoding="utf-8") as f:
            for entry in ds:
                f.write(json.dumps(entry, ensure_ascii=False) + "\n")
        
        print(f"✅ Saved to: {save_path} (Total: {len(ds)} items)")
    except Exception as e:
        print(f"❌ Failed to download {dataset_id}: {e}")

# if __name__ == "__main__":
#     for d_id in dataset_ids:
#         save_dataset_locally(d_id)
```

```python
import os
from datasets import load_dataset

# 1. 運営指定の全データセットIDリスト
dataset_ids = [
    "u-10bei/structured_data_with_cot_dataset_512_v2",   # 1-1 (標準)
    "u-10bei/structured_data_with_cot_dataset_512_v4",   # 1-2
    "u-10bei/structured_data_with_cot_dataset_512_v5",   # 1-3
    "u-10bei/structured_data_with_cot_dataset_512",      # 1-4
    "u-10bei/structured_data_with_cot_dataset_v2",       # 1-5
    "u-10bei/structured_data_with_cot_dataset",          # 1-6
    "daichira/structured-3k-mix-sft",                   # 2-1
    "daichira/structured-5k-mix-sft",                   # 2-2
    "daichira/structured-hard-sft-4k",                  # 2-3
    "u-10bei/dpo-dataset-qwen-cot"                      # DPO用
]

os.makedirs("data/hf_datasets", exist_ok=True)

for d_id in dataset_ids:
    name = d_id.split("/")[-1]
    print(f"📥 Downloading {d_id}...")
    try:
        ds = load_dataset(d_id, split="train")
        save_path = f"data/hf_datasets/{name}.jsonl"
        ds.to_json(save_path, force_ascii=False)
        print(f"✅ Saved: {save_path} ({len(ds)} items)")
    except Exception as e:
        print(f"❌ Error {d_id}: {e}")
```

## 中身を確認

```python
import json

def peek_jsonl(filename, n=1):
    path = f"data/hf_datasets/{filename}.jsonl"
    print(f"\n===== {filename} =====")
    with open(path, 'r', encoding='utf-8') as f:
        for i, line in enumerate(f):
            if i >= n: break
            data = json.loads(line)
            # 読みやすいように messages を整形して表示
            for msg in data.get('messages', []):
                print(f"【{msg['role'].upper()}】")
                # 長すぎると見づらいので先頭300文字くらい
                print(msg['content'][:300] + ("..." if len(msg['content']) > 300 else ""))

# v2 と Hard を比較
peek_jsonl("structured_data_with_cot_dataset_512_v2")
peek_jsonl("structured-hard-sft-4k")
```
